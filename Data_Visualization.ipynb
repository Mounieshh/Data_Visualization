```python
import matplotlib.pyplot as plt
x1 = [89,34,56,75,65,92,24]
y1 = [26,34,90,47,56,89,11]

x2 = [11,32,56,67,54,77,84]
y2 = [26,39,92,48,53,89,11]

plt.scatter(x1,y1,c='green',linewidths = 2,marker = "s",edgecolors="yellow",s = 50)
plt.scatter(x2,y2,c='red',linewidths = 2,marker = "o",edgecolors="blue",s = 30)

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
import numpy as np
x = np.array([3,4,5,6])
y = x*2

''' x = [2,3,5,6]
y = [5,7,8,9]
 '''
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("line graph")
plt.figure()

x1 = [2,3,5,6]
y1 = [5,7,8,9]
plt.plot(x1,y1,'-.',c= "indigo")
plt.title("New graph")
plt.show()
```

```python
x = np.array([1,2,3,4])
y = x*2
plt.plot(x,y)
x1 = [2,4,6,8]
y1 = [3,5,7,9]

plt.plot(x,y1,'-.')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Multiple plots")
plt.fill_between(x,y,y1,color= 'red',alpha = 0.2)
```

```python
x = np.array([80, 85, 90, 95, 100, 105])
y = np.array([240, 250, 260, 270, 280, 290])

plt.title("Grid Chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.plot(x, y)


plt.grid(axis = 'y',linestyle = '-.',linewidth = 3)

plt.show()
```
